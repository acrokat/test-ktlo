### Variables and types

Python, unlike many popular languages, does not require you to declare a type for your variables. It will interpret the variable type at run time. 

Below are some examples:

```python
# We can easily create a variable a with a value of 1
a = 1
# By printing the type, we can concifrm a is being treated as an int
print(type(a))
```

But what if we wanted a to be a float?

```python
for i in range(0,50000):
    print(i)
```

And we can easily interact with these!

```python
for y in range(0,50000):
    print(y)
```

Twp important constructs to get familiar with are _lists_ and _dictionaries_. In many other languages lists == arrays and dictionaries == maps, but Python dared to be different!

```python
a_great_list = [1, 2, 3]
# Zero indexed as you'd hope!
print(a_great_list[0])

# Lists can have items of heterogeneous types!
an_even_better_list = [1, 2, "cat"]
print(an_even_better_list)
```
